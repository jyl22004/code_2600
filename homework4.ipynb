Sample Mean: 78.4  
Standard Deviation: 10.782   
Confidence Interval: [70.687, 86.1129]

```python
import statistics

L1 = [65, 70, 68, 73, 75, 80, 85, 78, 90, 100]
mean = sum(L1) / len(L1)
std_dev = statistics.stdev(L1)

r = 2.262 * (std_dev / (len(L1) ** 0.5))

print(f'Sample Mean: {mean}')
print(f'Standard Deviation: {std_dev}')
print(f'Confidence Interval: [{mean - r}, {mean + r}]')
```

### 1b.
Bootstrap Means: [81.6, 75.4, 75.1, 83.2, 75.1, 73.0, 77.1, 79.9, 79.4, 78.0, 73.4, 75.5, 76.0, 77.8, 75.7, 76.3, 77.8, 83.6, 83.3, 77.5, 80.4, 79.1, 78.0, 75.9, 73.8, 81.4, 82.4, 79.5, 82.5, 77.5, 74.6, 84.6, 79.6, 75.0, 78.7, 80.4, 78.8, 79.4, 79.4, 77.3]  
Bootstrap Confidence Interval: [73.39, 83.625]

```python
import numpy as np

L2 = np.array([65, 70, 68, 73, 75, 80, 85, 78, 90, 100])

bootstrap_means = []
for _ in range(40):
    sample = np.random.choice(L2, size=len(L2), replace=True)
    bootstrap_means.append(np.mean(sample))

bootstrap_means = np.array(bootstrap_means)

# Estimate 95% CI from the bootstrap distribution
lower = np.percentile(bootstrap_means, 2.5)
upper = np.percentile(bootstrap_means, 97.5)

print(f'Bootstrap Means: {bootstrap_means.tolist()}')
print(f"Bootstrap Confidence Interval: [{lower}, {upper}]")
```

### 1c.
I would choose the method in part a to estimate the confidence interval for this case. The bootstrap method seems to be unnecessarily complex in this case. Part a only require using a few formulas, but part b also requires making bootstrapped samples in its calculations. While bootstrapping does yield a tighter confidence interval, I believe that the wider interval which the method from part a gives is more representative of the sample, which is relatively small in size and has a rather wide spread of scores. In this case, part a seems more apt.

### 3j.
The top 40% of customers have lift over 1.5 before lift proceeds to steadily decline, so it might be most effective to target this group of customers to maximize sales for cost.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Combine true labels and predicted probabilities
y_prob = rf_i.predict_proba(X_test)[:, 1]
lift_df = pd.DataFrame({'y_true': y_test, 'y_prob': y_prob})
lift_df['y_true'] = (lift_df['y_true']).astype(int)

# --- Sort and create deciles ---
lift_df = lift_df.sort_values('y_prob', ascending=False).reset_index(drop=True)
lift_df['decile'] = pd.qcut(lift_df.index, 10, labels=False, duplicates='drop')

# --- Aggregate by decile ---
decile_summary = lift_df.groupby('decile', observed=False).agg(
    total=('y_true', 'count'),
    positives=('y_true', 'sum')
).reset_index()

# --- Compute cumulative gains ---
total_positives = decile_summary['positives'].sum()
decile_summary['cumulative_positives'] = decile_summary['positives'].cumsum()
decile_summary['cumulative_gain'] = decile_summary['cumulative_positives'] / total_positives
decile_summary['data_fraction'] = (decile_summary.index + 1) / len(decile_summary)

# --- Compute lift ---
decile_summary['lift'] = decile_summary['cumulative_gain'] / decile_summary['data_fraction']

plt.figure(figsize=(8,6))
plt.plot(np.arange(1,11), decile_summary['lift'], marker='o')
plt.xlabel('Decile (10% segments)')
plt.ylabel('Lift')
plt.title('Lift Chart for Random Forest Model')
plt.grid(True)
plt.show()
```

### 3m.
Gradient boosting tree seems to be slightly better at segmenting the data. Gradient boosting maintains higher lift for the top 40% of customers, so if were to target the top 40% of customers for our marketing, we would expect gradient boosting to net more sales than the random forest.

```python
rf_probs = rf_i.predict_proba(X_test)[:, 1]
gb_probs = gb_l.predict_proba(X_test)[:, 1]

y_true_bin = (y_test).astype(int)

def lift_table(y_true, y_prob, n_bins=10):
    df = pd.DataFrame({'y_true': y_true, 'y_prob': y_prob})
    df = df.sort_values('y_prob', ascending=False).reset_index(drop=True)
    df['decile'] = pd.qcut(df.index, n_bins, labels=False, duplicates='drop')
    decile_summary = df.groupby('decile', observed=False).agg(
        total=('y_true', 'count'),
        positives=('y_true', 'sum')
    ).reset_index()
    total_pos = decile_summary['positives'].sum()
    decile_summary['cumulative_positives'] = decile_summary['positives'].cumsum()
    decile_summary['cumulative_gain'] = decile_summary['cumulative_positives'] / total_pos
    decile_summary['data_fraction'] = (decile_summary.index + 1) / len(decile_summary)
    decile_summary['lift'] = decile_summary['cumulative_gain'] / decile_summary['data_fraction']
    return decile_summary

rf_lift = lift_table(y_true_bin, rf_probs)
gb_lift = lift_table(y_true_bin, gb_probs)

plt.figure(figsize=(8,6))
plt.plot(rf_lift['data_fraction'], rf_lift['lift'], marker='o', label='Random Forest')
plt.plot(gb_lift['data_fraction'], gb_lift['lift'], marker='s', label='Gradient Boosting')
plt.xlabel('Proportion of Test Data (Deciles)')
plt.ylabel('Lift')
plt.title('Dual Lift Chart: Random Forest vs Gradient Boosting')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def plot_double_lift_chart(y_true, y_pred_model1, y_pred_model2,
                           model1_name='Model A', model2_name='Model B',
                           n_bins=10):
    """
    Plots a double lift chart comparing two regression models, 
    where bins are sorted by the ratio of Model 1 to Model 2 predictions.

    Parameters:
    - y_true: Array-like of true continuous values.
    - y_pred_model1: Array-like of predicted values from Model 1.
    - y_pred_model2: Array-like of predicted values from Model 2.
    - model1_name, model2_name: Names to display in the legend.
    - n_bins: Number of quantile bins to use for the lift chart.
    """

    # Build combined dataframe
    data = pd.DataFrame({
        'y_true': y_true,
        'pred1': y_pred_model1,
        'pred2': y_pred_model2
    })

    # Avoid division by zero
    data = data[data['pred2'] != 0]

    # Compute ratio of predictions
    data['ratio'] = data['pred1'] / data['pred2']

    # Sort descending by ratio
    data = data.sort_values(by='ratio', ascending=False)

    # Create quantile bins based on the ratio
    data['bin'] = pd.qcut(data['ratio'], n_bins, labels=False, duplicates='drop')

    # Overall means for normalization
    overall_mean_true = data['y_true'].mean()
    overall_mean_pred1 = data['pred1'].mean()
    overall_mean_pred2 = data['pred2'].mean()

    # Aggregate lift stats for each bin
    lift_data = data.groupby('bin').agg(
        actual_mean=('y_true', 'mean'),
        model1_mean=('pred1', 'mean'),
        model2_mean=('pred2', 'mean')
    ).reset_index()

    # Compute model error in each bin
    lift_data['model1_lift'] = lift_data['model1_mean'] / lift_data['actual_mean']
    lift_data['model2_lift'] = lift_data['model2_mean'] / lift_data['actual_mean']

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(lift_data['bin'], lift_data['model1_lift'], marker='s', linestyle='-', linewidth=2, label=f'{model1_name} Predicted Lift')
    plt.plot(lift_data['bin'], lift_data['model2_lift'], marker='^', linestyle='--', linewidth=2, label=f'{model2_name} Predicted Lift')

    plt.axhline(y=1, color='red', linestyle=':', label='Baseline (Lift = 1)')
    plt.title(f'Double Lift Chart by Ratio of {model1_name} / {model2_name}', fontsize=14)
    plt.xlabel(f'Bins (sorted by {model1_name} / {model2_name} ratio)')
    plt.ylabel('Lift (Mean / Overall Mean)')
    plt.xticks(np.arange(0, lift_data.shape[0]))
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.tight_layout()
    plt.show()

y_pred_rf = rf_i.predict(X_test)
y_pred_gb = gb_l.predict(X_test)
plot_double_lift_chart(y_test, y_pred_rf, y_pred_gb, 'Random Forest', 'Gradient Boosting')
```
